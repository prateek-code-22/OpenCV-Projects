```python
#Importing libraries:

import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Importing image to img.

img = cv2.imread(r'C:\Users\ANMOL\Desktop\DATA\MINI Projects\ML\OpenCV project\images\cr.jpg')


#to view same image use cvtColor to change image from BGR to RGB.

fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)


#display the original image.

plt.imshow(fix_img)
```

```python
#set the eye classifier to detect eyes in image.

eye_classifier = cv2.CascadeClassifier(r'C:\Users\ANMOL\Desktop\DATA\MINI Projects\ML\haarcascade\haarcascade_eye.xml')
```

```python
# detect_eyes()
# copy original image(fix_img) to copy_img.
# add classifier
# rectangle the eyes with red color.
# return processed image(copy_img)
```

```python
def detect_eyes(fix_img):
    copy_img = fix_img.copy()
    
    eyes_rects = eye_classifier.detectMultiScale(image=copy_img, scaleFactor=1.5,minNeighbors=3)
    
    for (x,y,w,h) in eyes_rects:
        cv2.rectangle(copy_img,
                     (x,y),
                     (x+w,y+h),
                     (255,0,0),4)
        
    return copy_img
```

```python
# result stores the processed image(copy_img) after passing original image(fix_img) as argument in detect_eyes function.

result = detect_eyes(fix_img)


# display the image.

plt.imshow(result)
```
