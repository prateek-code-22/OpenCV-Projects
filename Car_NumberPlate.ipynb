```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r'./images/car2.jpg')
```

```python
def display(img):
    fig = plt.figure(figsize=(10,8))
    ax= fig.add_subplot(111)
    new_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    ax.imshow(new_img)
```

```python
display(img)
```

```python
plate_classifier = cv2.CascadeClassifier(r'C:\Users\ANMOL\Desktop\DATA\MINI Projects\ML\Haarcascade\haarcascade_russian_plate_number.xml')
```

```python
def detect_plate(img):
    plate_img = img.copy()
    
    plate_rect = plate_classifier.detectMultiScale(image=plate_img, scaleFactor=5, minNeighbors=1)
    
    for (x,y,w,h) in plate_rect:
        cv2.rectangle(plate_img,
                     (x,y),(x+w,y+h),
                     (255,0,0),5)
        
    return plate_img
```

```python
result = detect_plate(img)
```

```python
display(result)
```
