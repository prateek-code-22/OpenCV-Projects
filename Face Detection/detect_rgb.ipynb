# DETECTING RBG PART IN AN IMAGE

```python
# importing package/libraries
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
pil = Image.open(r".\images\opencv.jpg")
pil
```

```python
# Converting image to numpy array
arr = np.array(pil)
arr
```

```python
#to get demension of matrix(M X N) and 3 is the thickness or layers.
arr.shape   
```

```python
#display image 
plt.imshow(arr) 
```

### Slicing and detecting red,green and blue part in image.
#### 0 is used for RED
#### 1 is used for BLUE 
#### 2 is used for GREEN.

```python
red = arr[:,:,0] 
plt.imshow(red)
```

```python
green = arr[:,:,1]
plt.imshow(green)
```

```python
blue = arr[:,:,2]
plt.imshow(blue)
```
