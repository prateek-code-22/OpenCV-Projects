```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
face_classifier = cv2.CascadeClassifier(r'C:\Users\ANMOL\Desktop\DATA\MINI Projects\ML\Haarcascade\haarcascade_frontalface_default.xml')
```

```python
img = cv2.imread('images\cr.jpg')
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
def detect_face(fix_img):
    face_rects = face_classifier.detectMultiScale(fix_img,1.1,2)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(fix_img,
                     (x,y), 
                     (x+w,y+h),
                     (255,0,0),7)
        
        return fix_img
```

```python
result = detect_face(fix_img)
plt.imshow(result)
```
