```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv2
```

```python
aircraft_classifier = cv2.CascadeClassifier(r"C:\Users\ANMOL\Desktop\DATA\MINI Projects\ML\Haarcascade Classifiers\Aircraft_2.xml")
helicopter_classifier = cv2.CascadeClassifier(r"C:\Users\ANMOL\Desktop\DATA\MINI Projects\ML\Haarcascade Classifiers\helikopter_1.xml")
```

```python
img = cv2.imread(r"C:\Users\ANMOL\Desktop\DATA\MINI Projects\ML\OpenCV project\images\copter.png")
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img)
```

```python
air = aircraft_classifier.detectMultiScale(gray_img)
```

```python
for (x,y,w,h) in air:
    cv2.rectangle(gray_img,(x,y),(x+w,y+h),(0,0,255),2)
    
    main_gray = gray_img[y:y+h, x:x+w]
    main_color = img[y:y+h, x:x+w]
    
    heli = helicopter_classifier.detectMultiScale(main_gray)
    
    for (xx,yy,ww,hh) in heli:
        cv2.rectangle(main_color,(xx,yy),(xx+ww,yy+hh),(255,0,0),2)
        
plt.imshow(gray_img)
```
