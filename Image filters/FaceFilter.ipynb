# 10 DIFFERENT FILTERS USING CVTCOLOR FUNCTION ON LIVE CAMERA.

```python
import cv2
cap =cv2.VideoCapture(0)

inp = int(input("Enter any number between 1 to 10"))
if inp == 1:
    while cap.isOpened():
        ret,frame = cap.read()
        if ret:
            hm = cv2.cvtColor(frame,cv2.COLOR_BGR2Luv)
            cv2.imshow("MyFace",hm)
        if cv2.waitKey(5) == ord('q'):
            break
        
elif inp == 2:
    while cap.isOpened():
        ret,frame = cap.read()
        if ret:
            hm = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
            cv2.imshow("MyFace",hm)
        if cv2.waitKey(5) == ord('q'):
            break
            
elif inp == 3:
    while cap.isOpened():
        ret,frame = cap.read()
        if ret:
            hm = cv2.cvtColor(frame,cv2.COLOR_BGR2HLS)
            cv2.imshow("MyFace",hm)
        if cv2.waitKey(5) == ord('q'):
            break
            
elif inp == 4:
    while cap.isOpened():
        ret,frame = cap.read()
        if ret:
            hm = cv2.cvtColor(frame,cv2.COLOR_BGR2RGBA)
            cv2.imshow("MyFace",hm)
        if cv2.waitKey(5) == ord('q'):
            break
            
elif inp == 5:
    while cap.isOpened():
        ret,frame = cap.read()
        if ret:
            hm = cv2.cvtColor(frame,cv2.COLOR_BGR2YCrCb)
            cv2.imshow("Myface",hm)
        if cv2.waitKey(5) == ord('q'):
            break
            
elif inp == 6:
    while cap.isOpened():
        ret,frame = cap.read()
        if ret:
            hm = cv2.cvtColor(frame,cv2.COLOR_BGR2YUV)
            cv2.imshow("Myface",hm)
        if cv2.waitKey(5) == ord('q'):
            break
            
elif inp == 7:
    while cap.isOpened():
        ret,frame = cap.read()
        if ret:
            hm = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
            cv2.imshow("Myface",hm)
        if cv2.waitKey(5) == ord('q'):
            break
            
elif inp == 8:
    while cap.isOpened():
        ret,frame = cap.read()
        if ret:
            hm = cv2.cvtColor(frame,cv2.COLOR_BGR2YCrCb)
            cv2.imshow("Myface",hm)
        if cv2.waitKey(5) == ord('q'):
            break
            
elif inp == 9:
    while cap.isOpened():
        ret,frame = cap.read()
        if ret:
            hm = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
            cv2.imshow("Myface",hm)
        if cv2.waitKey(5) == ord('q'):
            break
            
elif inp == 10:
    while cap.isOpened():
        ret,frame = cap.read()
        if ret:
            hm = cv2.cvtColor(frame,cv2.COLOR_BGR2Lab)
            cv2.imshow("Myface",hm)
            
        if cv2.waitKey(5) == ord('q'):
            break
        
else:
    print("program end")
    
cap.release()
cv2.destroyAllWindows()
```
